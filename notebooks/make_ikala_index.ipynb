```python
import csv
import os
import glob
import json
```

```python
lyrics_files = glob.glob("/Users/rachelbittner/.mir_datasets/iKala/Lyrics/*.lab")
```

```python
track_ids = sorted([os.path.basename(f).split('.')[0] for f in lyrics_files])
```

```python
track_ids
```

```python
with open("/Users/rachelbittner/.mir_datasets/iKala/id_mapping.txt", 'r') as fhandle:
    reader = csv.reader(fhandle, delimiter='\t')
    singer_map = {}
    for line in reader:
        if line[0] == 'singer':
            continue
        singer_map[line[1]] = line[0]
```

```python
ikala_index = {k: {} for k in track_ids}
```

```python
for key in ikala_index.keys():
    songid = key.split('_')[0]
    ikala_index[key]['singer_id'] = singer_map[songid]
    ikala_index[key]['song_id'] = key.split('_')[0]
    ikala_index[key]['section'] = key.split('_')[1]
    ikala_index[key]['audio_path'] = "iKala/Wavfile/{}.wav".format(key)
    ikala_index[key]['pitch_path'] = "iKala/PitchLabel/{}.pv".format(key)
    ikala_index[key]['lyrics_path'] = "iKala/Lyrics/{}.lab".format(key)
```

```python
with open("../mir_datasets/loaders/indexes/ikala_index.json", 'w') as fhandle:
    json.dump(ikala_index, fhandle, indent=2)
```
